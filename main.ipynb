# IMDB reviews - sentiment analysis

#### Contents:
1. Load and preprocess data
2. Train model
3. Get the results and accuracies

## 1. Load and preprocess data

```python
import re
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

### Load 

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
print(len(df_train))
print(len(df_test))
```

### Preprocess

```python
def preprocess_column(df, column_name):
    ps = PorterStemmer()
    texts_list = []
    for i, row in df.iterrows():
        text = row[column_name]
        text = re.sub(r'[^A-Za-z1-9]', ' ', text)
        text = str.lower(text)
        text = text.split()

        my_stopwords = stopwords.words('english')
        my_stopwords.remove("not")
        my_stopwords.append("br")

        stemmed_words = [ps.stem(word) for word in text if word not in my_stopwords]
        text = ' '.join(stemmed_words)
    
        texts_list.append(text)
    
    return texts_list
```

```python
train_texts = preprocess_column(df_train, "text")
test_texts = preprocess_column(df_test, "text")
```

## 2. Train models

### Vectorize texts

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 4300)
all_texts = train_texts + test_texts

X = cv.fit_transform(all_texts).toarray()

X_train = X[:len(train_texts),:]
X_test = X[len(train_texts):,:]

y_train = df_train["label"].values
y_test = df_test["label"].values
```

### Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

### Perceptron

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python
ann_model = Sequential()

ann_model.add(Dense(100, activation='relu'))
ann_model.add(Dense(units=100, activation='relu'))
ann_model.add(Dense(units=100, activation='relu'))
ann_model.add(Dense(1, activation='sigmoid'))

ann_model.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy'])
ann_model.fit(X_train, y_train, epochs=1000)
```

## 3. Get the results and accuracies

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

### Naive Bayes model

```python
nb_pred = nb_model.predict(X_test)
nb_accuracy = accuracy_score(nb_pred, y_test)
nb_conf = confusion_matrix(nb_pred, y_test)
```

```python
print("Accuracy:", nb_accuracy)
print("\n")
print("Confusion matrix:\n", nb_conf)
```

### Logistic regression

```python
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(lr_pred, y_test)
lr_conf = confusion_matrix(lr_pred, y_test)
```

```python
print("Accuracy:", lr_accuracy)
print("\n")
print("Confusion matrix:\n", lr_conf)
```

### Perceptron

```python
ann_pred = [round(x[0]) for x in ann_model.predict(X_test)]
ann_accuracy = accuracy_score(ann_pred, y_test)
ann_conf = confusion_matrix(ann_pred, y_test)
```

```python
print("Accuracy:", ann_accuracy)
print("\n")
print("Confusion matrix:\n", ann_conf)
```
